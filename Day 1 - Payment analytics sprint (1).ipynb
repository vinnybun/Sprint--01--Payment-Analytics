## Day 1 - THINKING LIKE A FINTECH DATA SCIENTIST
Data doesn't solve the problem; questions do.
As a professional data scientist, you need to ask:
- What problems are we solving?
- Who is affected?
- How do we measure success?
- What data do we need?
- What decisions will this analysis support?

After a clear understanding of the problem by creating the answers to this questions, We then follow the data science workflow :
- Understanding the Business Problem.
- Define success metrics (KPIs).
- Collect and understand the data.
- Clean and prepare the data.
- Explore the data (EDA).
- Build analytical and predictive models.
- Communicate insights.
- Deploy and monitor solutions.

For day 1, we would be focusing on the first three elements of the data science workflow, with the aim of understanding the data.

```python
#import library
import pandas as pd

#load CSV file 
df = pd.read_csv("Synthetic_Financial_datasets_log.csv")

#display the first five rows
print("The first five rows ", "\n",  df.head(), "\n")


#Display the last five rows
print("The last five rows ", "\n",  df.tail())
```

```python
#print the shape of the 
print("The shpae of the dataframe is : ", df.shape)
```

```python
#list of the column names
columns = list(df.columns)
print("The columns of the dataset are:", "\n")
for col in columns:
    
    print(col)

```

```python
#display the data types
dtype = df.dtypes
print(dtype)
```

```python
#count of missing values
missing_values = df.isnull().sum()
print("The count of missing values for each column ")
print(missing_values)
```

```python
#summary statistics of numerical columns
num_stat = df.describe().astype(int).T
print(num_stat)
```
